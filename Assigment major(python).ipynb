3. Does the data show any trends in terms of the number of honey
producing colonies and yield per colony before 2006, which was when
concern over Colony Collapse Disorder spread nationwide?

```python
plt.figure(figsize=(15, 7))
sns.pointplot(x='year', y='yieldpercol', data=colonybf_2006, estimator=sum, ci=None)
plt.title('Yield per colony before 2006' ,color='maroon')
plt.xticks(rotation=90 , size=15) 
plt.yticks( size=15) 
plt.show()
```

```python
plt.figure(figsize = (25,10))

sns.barplot(x="year", y="prodvalue", data=df.sort_values("totalprod", ascending=False),
            label="prodction value",color="r",  ci=None, edgecolor='0',linewidth=4)

sns.barplot(x="year", y="totalprod", data=df.sort_values("totalprod", ascending=False),
            label="Total Production",color="y",  ci=None,alpha=0.7)
plt.legend(ncol=2, loc="upper left", frameon=True, fontsize=17)
plt.title('The total honey production & value of production in every year',color='blue')
plt.show()
```
